### Plus de reprsentations graphiques
MATPLOTLIB permet de reprsenter graphiquement des sries de nombres que nous lui fournissons **explicitement**.

Dans le cas prcdent, *x_data* et *y_data* pour les valeurs des points en abscisse et ordonne respectivement.

TOUT graphique doit dbuter par un appel  la mthode *subplots* de l'objet *matplotlib.pyplot* (ici nomm *plt*).

`fig, ax = plt.subplots()`
- `fig` est un objet [Figure](https://matplotlib.org/stable/api/figure_api.html) qui controle les paramtres de toute la figure
- `ax` est **un ou des** objets [Axes](https://matplotlib.org/stable/api/axes_api.html#basic), **chaque objet "Axes" controlant un seul graphique**

#### Superposer plusieurs graphiques

Les appels successifs au mme objet Axes pour diffrent graphiques vont les superposer.
En reprenant l'exemple prcdent, dessinons deux nuages de points sur le mme graphique.
```python
x1_data = np.random.normal(1, 0.2, 50)
y1_data = np.random.normal(0, 0.2, 50)

x2_data = np.random.normal(0.5, 0.2, 50)
y2_data = np.random.normal(-3, 0.1, 50)
```
Puis dessinons les successivement sur le **mme objet "Axes"**.

$\rightarrow$ Profitons en pour dcouvrir
* le paramtre *marker* de la fonction *Axes.scatter*
* la mthode *Figure.legend()* en combinaison avec le paramtre *label*

(https://matplotlib.org/stable/api/markers_api.html?highlight=marker#module-matplotlib.markers)
```python
fig2, ax2 = plt.subplots(figsize=(4,4))
ax2.scatter(x1_data, y1_data, label="dataset1")
ax2.scatter(x2_data, y2_data, marker="X")

fig2.legend(loc='upper left')
```

### Reprsentation de fonctions continues

Ces graphiques sont produits par la mthode [Axes.plot](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html?highlight=plot#matplotlib.axes.Axes.plot)

Elles necessite un ensemble discret de valeur de x et y, la fonction plot se chargeant d'interpoler la courbe entre chaque valeur. On devra donc:
- gnrer un ensemble de valeurs dans le domaine de la fonction
- appliquer la fonction  chacune de ces valeurs pour produire le "co-domaine"
- passer les listes de valeurs du domaine et du co-domaine  *Axes.plot*


#####  Sries de valeurs galement rparties entre deux bornes  l'aide de [np.linspace](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html)

Cette fonction numpy est essentielle pour reprsenter le domaine des fonctions continues. Par exemple, gnrons 1000 valeurs dans l'intervalle \[-10;10\].

```python
x  = np.linspace(-10, 10, 10000)
```
Puis appliquons  "ce domaine" les fonctions *cos(x)* et sin(x).
```python
y1 = np.cos(x)
y2 = np.sin(x)
```

##### Grille de graphiques
L'organisation de plusieurs graphiques dans une figure est controle par les paramtre *ncols* et "nrows".

```python
fig3, ax3 = plt.subplots(figsize=(8,4),nrows = 1, ncols = 2)
````

Puis dessinons deux graphiques dans la mme figure:
- Le dessin de gauche (*ax3[0]*) tant la fonction cos(x), entre -10 et 10 en vert pais.
- Le dessin de droite (*ax3[1]*) tant la fonction sinx(x) entre -10 et 10 en traits d'union.

```python
fig3, ax3 = plt.subplots(figsize=(8,4),nrows = 1, ncols = 2)
ax3[0].plot(x, y1, color='green', linewidth=2)
ax3[1].plot(x, y2, linestyle='dashed')
```

<h5>PUIS</h5>

<span style="color:firebrick;font-weight:bold"> $\rhd$ Rajouter une [grille](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.grid.html) au graphique de droite et le titre de la fonction trigonometrique  chaque graphique.</span>

`````python
x  = np.linspace(-15, 15, 10000)
y1 = np.cos(x)
y2 = np.sin(x)
y3 = np.arctan(x)
y4 = np.tan(x)

fig3, ax3 = plt.subplots(figsize=(8,4),nrows = 2, ncols = 2)

ax3[0][0].plot(x, y1, color='green', linewidth=1)
ax3[0][0].set_title("Cosinus")

ax3[0][1].plot(x, y2, linestyle='dashed', linewidth=2)
ax3[0][1].grid()
ax3[0][1].set_title("Sinus")

ax3[1][0].plot(x, y3, color='purple', linewidth=3, linestyle='dashdot')
ax3[1][0].set_title("Arctan")

ax3[1][1].plot(x, y4, color='orange', linewidth=4, linestyle='dotted')
ax3[1][1].set_title("Tan")
`````
